```python
from scrna.datasets import *
from scrna.prc import *
from scrna.rmt import *
import matplotlib.pyplot as plt
import numpy as np
from scrna.utils import experiment_repr, concat_pipelines
```

```python
# In the scLENS paper the tot counts in the normalization is set to 1 instead of 10000
rng = np.random.default_rng(12)
ngenes = 2500
for dataset, info in DATASETS.items():

    print('processing dataset', dataset)
    large, small, _, small_pipeline = get_subsampled_datasets(dataset, rng=rng)

    pr_pipeline = get_preprocessing_pipeline(ngenes = ngenes)
    
    normlog = Pipeline(steps=[('norm', SizeNormalizer(use_obs='size_factor')),
                              ('log', LogNormalizer())], verbose=True)
    small = normlog.fit_transform(pr_pipeline.fit_transform(small))

    spec = {
        'trunk': concat_pipelines([small_pipeline, pr_pipeline, normlog]),
        'name_trunk': info['name']
    }
    experiment_repr(spec, 'figures/esd_%s_%d' %  (dataset, ngenes))
    sep_cov_est = SeparableCovarianceEstimator(epsilon=5e-4).fit(small)
    fig, ax = sep_cov_est.plot_eigenspectrum(small)

    fig.savefig('figures/esd_sep_%s_%d.pdf' % (dataset, ngenes), dpi=200)

    smallbw = BiwhiteningScaler(with_mean=False).fit_transform(small.copy())
    bw_cov_est = BiwhitenedCovarianceEstimator().fit(smallbw)
    fig, ax = bw_cov_est.plot_eigenspectrum(smallbw)

    fig.savefig('figures/esd_bw_%s_%d.pdf' % (dataset, ngenes), dpi=200)
    

plt.show()
```
